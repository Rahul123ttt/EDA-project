```python
data['family']=data['SibSp']+data['Parch']+1
data=data.drop(['SibSp','Parch'],axis=1)
data=data.drop('PassengerId',axis=1)
data=data.drop('Embarked',axis=1)
data
```

```python
x=data.drop('Survived',axis=1).values
y=data['Survived'].values
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)
```

```python
from sklearn.metrics import accuracy_score
```

```python
Ir=LogisticRegression()
Ir.fit(x_train,y_train)
Irpred=Ir.predict(x_test)
```

```python
accuracy_score(y_test,Irpred)
```

```python
from sklearn.model_selection import GridSearchCV
# creating a hyper parameter grid
c_space=np.logspace(-5,8,15)
param_grid={'C':c_space}
# instantiating the gridsearchingcv object
logreg_cv=GridSearchCV(Ir,param_grid,cv=5)

logreg_cv.fit(x_train,y_train)
# Print the tuned parameter and score
print("tuned logisticed Regression :  {}".format(logreg_cv.best_params_))
print("best score is {}".format(logreg_cv.best_score_))

```
